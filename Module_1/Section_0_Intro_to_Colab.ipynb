# Introduction to Colab

Colab is an interactive computational environment to write and test code through online web interface. Libraries and tools are already pre-installed in colab so you can run your codes without any required set up on your laptop. All you need is your web browser! In Colab, we'll use the programming language python. The Colab documents where you run your python codes are called jupyter notebooks with the file extention ipynb which stands for ipython notebook. The notebooks are made up of cells where you can write text and code as well as visualize graphs and images.
