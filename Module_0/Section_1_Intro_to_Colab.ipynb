# 1.0 | Introduction to Colab

Colab is an interactive computational environment to write and test code through online web interface. Libraries and tools are already pre-installed in colab so that you can run your code without any required set up on your laptop. All you need is your web browser! In Colab, we'll use the python programming language. The Colab documents where you run your python code are called jupyter notebooks with the file extention ipynb which stands for ipython notebook. The notebooks are made up of cells where you can write text and code as well as visualize graphs and images.

When you select cell, there are a few control options available on the right side. You can move the cells up or down using the arrows, get a link to a specific cell, delete the cell, and more.
Colab saves your changes automatically every few seconds. You can also save your changes manually by selecting File then Save from the menu options at the top of the notebook. There is even an option to save your work to Google Drive by selecting "Save a copy in Drive." You can download your jupyter notebook under the File menu as well. Use the .ipynb extension.


The code cells in your notebooks are stored by the so-called kernel within the limits of memory. Sometimes, you may want to restart your kernel to free the memory of the code cells you would have run so far. To free up memory, you can restart the kernel by using the menu at the top of the notebook. Click Runtime, and select Restart Runtime.
